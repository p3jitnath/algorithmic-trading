```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../05-Pandas-with-Time-Series/time_data/walmart_stock.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format="%Y-%m-%d")
```

```python
df.info()
```

```python
df = df.set_index('Date')

df.head()
```

```python
df.resample(rule='A').mean()
```

```python
df.resample(rule='BQ').mean()
```

```python
def first_day(entry):
    return entry[0]

df.resample(rule='A').apply(first_day)
```

```python
df['Close'].resample('M').mean().plot(kind='bar', figsize=(16, 6))
```
